```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color="blue", linestyle="--")
plt.show()
```

```python
categorias = ["A", "B", "C", "D"]
valores = [5, 7, 3, 8]

plt.bar(categorias, valores, color="orange")
plt.show()
```

```python
x = np.random.rand(300)
y = np.random.rand(300)

plt.scatter(x, y, color="purple", alpha=0.7, edgecolor="black")
plt.show()
```

```python
fatias = [30, 20, 25, 25]
labels = ["A", "B", "C", "D"]

plt.pie(fatias, labels=labels, autopct="%1.1f%%", startangle=90, colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"])
plt.title("Grfico de Pizza")
plt.show()
```
